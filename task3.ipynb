```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

Задание 3

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = np.power(x, 2)
```

```python
y2 = 2 * x + 0.5
```

```python
y3 = -3 * x - 1.5
```

```python
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax[0, 0].plot(x, y1)
ax[0, 0].set_title('График y1', color='DarkRed', family = 'serif', fontsize=10)
ax[0, 0].set_xlim([-5, 5])
ax[0, 1].plot(x, y2)
ax[0, 1].set_title('График y2', color='DarkRed', family = 'serif', fontsize=10)
ax[1, 0].plot(x, y3)
ax[1, 0].set_title('График y3', color='DarkRed', family = 'serif', fontsize=10)
ax[1, 1].plot(x, y4)
ax[1, 1].set_title('График y4', color='DarkRed', family = 'serif', fontsize=10)
```
