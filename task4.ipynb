```python
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
import warnings 
warnings.filterwarnings('ignore')
```

```python
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
dataset_path = r'd:\Документы\Учеба.GeekBrains\Numpy_Pandas\lesson4\creditcard.csv'
```

```python
df = pd.read_csv(dataset_path, sep = ',')
df.info()
```

```python
c_class = df['Class'].value_counts()
```

```python
c_class.plot(kind='bar')
plt.show()
```

```python
c_class.plot(kind='bar', logy=True)
plt.show()
```

```python
rcParams["figure.figsize"] =9, 12
cl0 = df.loc[df['Class'] == 0, 'V1']
cl1 = df.loc[df['Class'] == 1, 'V1']
plt.hist(cl0, bins=20, alpha=0.5, density=True, ec="black", color='Grey')
plt.hist(cl1, bins=20, alpha=0.5, density=True, ec="black", color='Red')
plt.legend(['Class 0', 'Class 1'])
plt.xlabel('V1')
plt.show()
```
