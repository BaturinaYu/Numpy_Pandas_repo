```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

Задание 2

```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
title_font = {'color': 'Grey', 'fontsize': 16, 'fontweight': 'bold', 'family': 'serif'}
plt.plot(t, f, label = 'cos(t)', color = 'Green')
plt.title('График f(t)', fontdict = title_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.xlabel('Значения t', fontsize = 12, family = 'serif', color = 'Grey')
plt.ylabel('Значения f', fontsize = 12, family = 'serif', color = 'Grey')
plt.legend(loc = 'lower right')
plt.show()
```
